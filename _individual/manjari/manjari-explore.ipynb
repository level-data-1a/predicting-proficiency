```python
import pandas as pd
import json


json_file_path = 'data/district.json'

with open(json_file_path, 'r') as json_file:
    data = json.load(json_file)

#normalize/ convert list to dict
def normalize_json_column(column):
    if isinstance(column, list) and len(column) > 0 and isinstance(column[0], dict):
        return pd.json_normalize(column, sep='_')
    return pd.DataFrame()  

# Normalize each column
benchmarks_df = normalize_json_column(data['benchmarks'])
schools_df = normalize_json_column(data['schools'])
course_sections_df = normalize_json_column(data['courseSections'])
course_section_rosters_df = normalize_json_column(data['courseSectionRosters'])
scores_df = normalize_json_column(data['scores'])

```

```python
import os

# Get the file size in bytes
file_size = os.path.getsize(json_file_path)

# Convert to KB, MB, GB as needed
file_size_kb = file_size / 1024
file_size_mb = file_size_kb / 1024

print(f"File size: {file_size} bytes")
print(f"File size: {file_size_kb:.2f} KB")
print(f"File size: {file_size_mb:.2f} MB")
```

```python
# Step 1: Get unique values, missing values, and count of zeros
for col in scores_df.columns:
    unique_values = scores_df[col].unique()  # Unique values
    num_missing = scores_df[col].isna().sum()  # Missing values (NaN)
    num_zeros = (scores_df[col] == 0).sum()  # Count of zeros
    
    print(f"Column: {col}")
    print(f"Unique values: {unique_values}")
    print(f"Missing values (NaN): {num_missing}")
    print(f"Number of zeros: {num_zeros}")
    print("-" * 40)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Convert 'value' column to numeric, forcing errors to NaN
benchmarks_df['value'] = pd.to_numeric(benchmarks_df['value'], errors='coerce')

# Boxplot to visualize the value distribution for each dataSource
plt.figure(figsize=(10, 6))
sns.boxplot(x='dataSource', y='value', data=benchmarks_df)
plt.xticks(rotation=90)
plt.title('Value Distribution Across Data Sources')
plt.show()

# Group by year and dataSource to visualize average values over time
df_grouped = benchmarks_df.groupby(['year', 'dataSource'])['value'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='year', y='value', hue='dataSource', data=df_grouped)
plt.title('Average Value Across Data Sources by Year')
plt.show()
```

```python
# Special Ed Subgroup Distribution in scores_df
sns.countplot(x='subgroup_specialEd', data=scores_df)
plt.title('Distribution of Special Education Students in Scores')
plt.show()

# Lunch Status Subgroup Distribution in scores_df
sns.countplot(x='subgroup_lunchStatus', data=scores_df)
plt.title('Distribution of Lunch Status in Scores')
plt.show()

# Bottom 25 Subgroup Distribution in scores_df
sns.countplot(x='subgroup_bottom25', data=scores_df)
plt.title('Distribution of Bottom 25% Students in Scores')
plt.show()
```

```python
# Define ACT score columns in scores_df
act_columns = ['value_actMathScore', 'value_actEnglishScore', 'value_actReadingScore', 'value_actScienceScore', 'value_actCompositeScore']

# Summary statistics for ACT scores in scores_df
act_stats = scores_df[act_columns].describe()
print(act_stats)
```

```python
# Visualize ACT score distributions in scores_df with adjusted x-axis
for column in act_columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(scores_df[column], kde=True, bins=15)  # Reduced number of bins to avoid cramming
    plt.title(f'Distribution of {column} in Scores', fontsize=14)
    plt.xlabel(f'{column}', fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    
    # Adjust the x-axis limits to focus on meaningful ranges (e.g., ACT scores are typically between 0-36)
    plt.xlim(0, 36)
    
    # Rotate the x-axis labels if needed
    plt.xticks(rotation=45)
    
    plt.tight_layout()  # Adjust layout for better spacing
    plt.show()
```
