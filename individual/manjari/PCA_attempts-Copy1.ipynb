```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

# Assuming your PCA dataframe is named 'pca_df'
# and has already been computed

# If not already computed, you can compute PCA like this:
# pca = PCA(n_components=5)
# pca_scores = pca.fit_transform(your_data)

# Create a PCA dataframe
#pca_df = pd.DataFrame(pca_scores, columns=['PC1', 'PC2', 'PC3', 'PC4', 'PC5'])

first_21_pcs_df
# Define a function to perform K-means clustering and plot the results
def plot_clusters(first_21_pcs_df, pc_x, pc_y, n_clusters):
    # Select the two principal components for the scatter plot
    data = first_21_pcs_df[[pc_x, pc_y]]

    # Perform K-means clustering
    kmeans = KMeans(n_clusters=n_clusters)
    kmeans.fit(data)
    first_21_pcs_df[f'cluster_{pc_x}_{pc_y}'] = kmeans.labels_

    # Plot the scatter plot with cluster coloring
    plt.figure(figsize=(6, 6))
    sns.scatterplot(x=pc_x, y=pc_y, hue=f'cluster_{pc_x}_{pc_y}', data= first_21_pcs_df, palette='Set1', legend='full')
    plt.title(f'K-means Clustering on {pc_x} vs {pc_y} with {n_clusters} clusters')
    plt.show()

# Example to find the appropriate number of clusters using the elbow method
def find_optimal_clusters(data, max_k):
    iters = range(1, max_k + 1)
    sse = []
    for k in iters:
        model = KMeans(n_clusters=k)
        model.fit(data)
        sse.append(model.inertia_)

    plt.figure(figsize=(6, 6))
    plt.plot(iters, sse, marker='o')
    plt.xlabel('Cluster Centers (k)')
    plt.ylabel('SSE')
    plt.title('Elbow Method for Optimal Clusters')
    plt.show()

# For example, to find optimal number of clusters for PC1 and PC2
find_optimal_clusters(first_21_pcs_df[['PC1', 'PC2']], max_k=10)

# After inspecting the elbow method plot, assume we decide to use 3 clusters for PC1 vs PC2
plot_clusters(first_21_pcs_df, 'PC1', 'PC2', n_clusters=3)

# Similarly, repeat for other scatter plots (PC1 vs PC3, PC2 vs PC3, etc.)
# Here's an example for PC1 vs PC3 with a hypothetical 4 clusters
plot_clusters(first_21_pcs_df, 'PC1', 'PC3', n_clusters=4)

# You can continue this for each pair (PC1 vs PC4, PC2 vs PC3, etc.)
```
